```python
def page_views(article, start, end):
    result = (pageviewapi.per_article('en.wikipedia', article, start, end,
                        access='all-access', agent='all-agents', granularity='daily'))
    views = [i["views"] for i in result["items"]]
    return views
    
```

```python
page_views('Armenia', '20180701', '20180731')
```

```python
pprint(pageviewapi.top('en.wikipedia', year = "2018", month = "08", day = "01", access='all-access'))
```

```python
def top_n(n, month, day):
    result = pageviewapi.top('en.wikipedia', year=2018, month=month, day=day, access='all-access')  #year=str(datetime.today())
    names = [i["article"] for i in result["items"][0]["articles"][:n]]
    return names

  
```

```python
top_n(n=10, month = "07", day = "15")
```

```python
def top_n_df(n, month, day):
    result = pageviewapi.top('en.wikipedia', year=2018, month=month, day=day, access='all-access')  #year=str(datetime.today())
    articles = result["items"][0]["articles"][:n]
    df = pd.DataFrame(articles)
    return df
```

```python
def dist_calc(loc1, loc2):
    url = "https://maps.googleapis.com/maps/api/directions/json?origin="+str(loc1)+"&destination="+str(loc2)
    response = requests.get(url)
    output = response.json()
    return output['routes'][0]["legs"][0]["distance"]["text"]
```

```python
dist_calc("Yerevan", "Tbilisi")
```
